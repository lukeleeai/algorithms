```python
def minimumJumps(forbidden, a, b, x):
    res = -1
    
    def run(pos, cnt, wentBack):
        nonlocal res
        
        if pos in forbidden or pos < 0:
            return
        elif wentBack and pos > x:
            return
        elif pos == x:
            return

        bck = run(pos-b, cnt+1, True) if not wentBack else 0
        if bck:
            res = bck
            return res
        fwd = run(pos+a, cnt+1, False)
                    
    run(0, 0, False)
    return res
```
